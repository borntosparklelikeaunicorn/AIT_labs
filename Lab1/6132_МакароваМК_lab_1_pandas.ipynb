```python
result = data.groupby("playlist_name")["tempo"].mean() #группируем данные по столбцу playlist_name и подсчитываем среднее значение в столбце tempo для каждой группы
result = result.reset_index() #сбрасываем индекс
result = result.rename(columns={'tempo': 'mean_tempo'}) #переименовываем столбец со ср. знач.
pivot_df = result.pivot(index='playlist_name', columns='mean_tempo', values='mean_tempo')
pivot_df
```

7. Вычислить квантили распределения какого-либо вещественного признака (с использованием numpy.quantile или numpy.percentile).

```python
duration_ms_column = data["duration_ms"] #извлекаем столбец duration_ms
q1 = np.percentile(duration_ms_column, 25) #вычисляем первый квартиль (25-й перцентиль)
median = np.percentile(duration_ms_column, 50) #вычисляем медиану (50-й перцентиль)
q3 = np.percentile(duration_ms_column, 75) #ычисляем третий квартиль (75-й перцентиль)

print("Первый квантиль:", q1)
print("Медиана:", median)
print("Третий квантиль:", q3)
```
